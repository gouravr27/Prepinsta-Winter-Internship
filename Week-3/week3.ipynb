```python
import pandas as pd
import numpy as np
```

###FIRST DATASET

```python
data = pd.read_csv("/content/Data-cleaning-for-beginners-using-pandas.csv")
data
```

AGE COLUMN

```python
avg_age = data['Age'].mean()
data['Age']=data.Age.fillna(avg_age)
data['Age']=data.Age.round(decimals=1)
```

```python
data['Age']=data['Age'].astype('int64')
data
```

SALARY COLUMN

```python
data['Salary']= data['Salary'].str.replace("$","")
data['Salary']= data['Salary'].str.replace("k","000")
data
```

RATING COLUMN

```python
avg_rating=data['Rating'].mean()
data['Rating']=data['Rating'].fillna(avg_rating)
data['Rating']=data.Rating.round(decimals=1)
data
```

```python
data['Rating']=data['Rating'].replace(-1,1)
data.head()
```

LOCATION COLUMN

```python
data['Location']= data['Location'].replace({"India,In" : "India" ,"India In": "India", "New York,Ny": "New York", "Australia Aus":"Australia"})
data.head()
```

ESTABLISHED COLUMN

```python
data['Established'] = data['Established'].replace(-1, np.nan)
```

```python
established = data['Established'].mean()
data['Established']=data['Established'].fillna(established)
data['Established']=data.Established.round(decimals=1)
data['Established']=data['Established'].astype('int64')
data
```

EASY APPLY COLUMN

```python
data['Easy Apply']= data['Easy Apply'].replace("-1","FALSE")
data
```

```python
data['Easy Apply'] = data['Easy Apply'].astype(bool)
```

DATA TYPE

```python
data.dtypes
```

##SECOND DATASET

```python
df = pd.read_csv("/content/chipotle.tsv",sep='\t')
df
```

```python
df.isna().sum()
```

```python
df['choice_description']=df['choice_description'].fillna('None')
df.head()
```

```python
df.isna().sum()
```

```python
df['item_price'] = df['item_price'].replace('$','')
df['item_price'] = df['item_price'].astype(int)
df.head()
```

```python
df['choice_description']=df['choice_description'].str.replace("[" ," ")
df['choice_description']=df['choice_description'].str.replace("]" ," ")
df['choice_description']
```

```python
df['choice_description'] =df['choice_description'].astype(str)
```

```python
df.describe()
```

Quantity-Item Relationship:

```python
quantity_item_relation = df.groupby('item_name')['quantity'].sum().reset_index(name= 'total_quantity')
quantity_item_relation
```

Quantity-price Relationship:

```python
quantity_item_relation = df.groupby('item_name')['item_price'].sum().reset_index(name= 'total_price')
quantity_item_relation
```

MOST ORDERED ITEMS

```python
most_order=df.sort_values(by="quantity",ascending=False)
most_order.head()
```

REVENUE OF EACH ITEMS

```python
df['total_price_calculated'] = df['quantity'] * df['item_price']
revenue_each = df[df['total_price_calculated'] != df['item_price']]
revenue_each
```

```python
unique_item_names = df['item_name'].unique()
unique_item_names
```

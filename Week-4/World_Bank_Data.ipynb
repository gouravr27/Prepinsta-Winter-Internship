```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snp
import plotly.express as px
```

##READING DATASET

```python
country= pd.read_csv("/content/Metadata_Country.csv")
population= pd.read_csv("/content/country_population.csv")
fertility= pd.read_csv("/content/fertility_rate.csv")
life_expectancy=pd.read_csv("/content/life_expectancy.csv")
```

##UPDATING

```python
country=country[['Country Code','Region']]
country.head()
```

```python
population.head()
```

```python
population.drop(['Country Name','Indicator Name','Indicator Code'],axis =1,inplace=True)
population.head()
```

```python
fertility.drop(['Country Name','Indicator Name','Indicator Code'],axis =1,inplace=True)
fertility.head()
```

```python
life_expectancy.drop(['Country Name','Indicator Name','Indicator Code'],axis =1,inplace=True)
life_expectancy.head()
```

##Handle Missing Values

```python
country.isna().sum()
```

```python
population.dropna(axis =0,inplace=True)
```

```python
fertility.dropna(axis =0,inplace=True)
```

```python
life_expectancy.dropna(axis =0,inplace=True)
```

##MELTING AND THEN MERGING DATA

```python
years = [str(i) for i in range(1960, 2017)]
```

```python
population = pd.melt(population,
        id_vars='Country Code',
        value_vars=years,
        var_name='Year',
        value_name='Population')
```

```python
population.head()
```

```python
fertility = pd.melt(fertility,
        id_vars='Country Code',
        value_vars=years,
        var_name='Year',
        value_name='Fertility Rate')
```

```python
fertility.head()
```

```python
life_expectancy = pd.melt(life_expectancy,
        id_vars='Country Code',
        value_vars=years,
        var_name='Year',
        value_name='life expectancy')
```

```python
life_expectancy.head()
```

```python
merge=pd.merge(country,population, how='left',on='Country Code')
```

```python
merge.head()
```

```python
merge=pd.merge(merge,fertility, how='left',on=['Country Code','Year'])
merge.head()
```

```python
merge=pd.merge(merge,life_expectancy, how='left',on=['Country Code','Year'])
merge.head()
```

```python
merge.dropna(axis=0, inplace=True)
```

```python
merge.isna().sum()
```

##GRAPHS

```python
merge.head()
```

```python
fig = px.bar(merge, x='Region', y='Population', color="Region",
  animation_frame="Year", animation_group="Country Code", range_y=[0,3000000000])
fig.update_layout(title="Year vs Population")
fig.show()
```

```python
fig_fertilty_life = px.scatter(merge, x="Fertility Rate", y="life expectancy", animation_frame="Year", animation_group="Country Code",
           size="Population", color="Region", hover_name="Country Code", size_max=100,
           range_x=[0,10],range_y=[10,100])
fig_fertilty_life.update_layout(title="Fertility Rate vs Life Expectancy")
```

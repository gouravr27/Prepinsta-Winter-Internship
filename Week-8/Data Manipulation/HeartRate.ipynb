##HeartRate Data

* Import Libraries

```python
import numpy as np
import pandas as pd
```

* Loading Dataset

```python
df=pd.read_csv("/content/heartrate_seconds_merged.csv")
df.head()
```

## Data Cleaning

* Now we split the **Time** column into two seperate colums **Activity_Date** and **Activity_Time**.

```python
df[['Activity_Date','Activity_Time']]= df['Time'].str.split(' ', 1, expand=True)
df.head()
```

* Now we will drop the **Time** Column as we split it, so we don't need this column now.

```python
df.drop(['Time'], axis=1)
```

* Now we will changed the column order

```python
df=df[['Id','Activity_Date','Activity_Time','Value']]
```

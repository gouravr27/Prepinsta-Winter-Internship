# Minute Dataset

## Import Libraries

```python
import numpy as np
import pandas as pd
```

## Loading Dataset

In minutes we have two type of datasets:
1.   Narrow (Every minute in a single row)
2.   Wide (Every minute in a single column) <br>
We will use narrow one

* First we will Load the Minute Step Csv File

```python
m1= pd.read_csv("/content/minuteStepsNarrow_merged.csv")
m1.head()
```

* Checking the null values

```python
m1.isnull().sum()
```

* Loading Minutes Calories file

```python
m2= pd.read_csv("/content/minuteCaloriesNarrow_merged.csv")
m2.head()
```

* Merging both steps and calories csv files

```python
merge = pd.merge(m1,m2,on=('Id','ActivityMinute'), how='left')
merge.head()
```

* Loading Minutes Intensity file

```python
m3= pd.read_csv("/content/minuteIntensitiesNarrow_merged.csv")
m3.head()
```

* Merging both the previous merges filed and intensity csv file

```python
merge1 = pd.merge(merge,m3,on=('Id','ActivityMinute'), how='left')
merge1.head()
```

* Loading Minutes METs file

```python
m4= pd.read_csv("/content/minuteMETsNarrow_merged.csv")
m4.head()
```

* Merging both the previous merges filed and METs csv file

```python
min_merge = pd.merge(merge1,m4,on=('Id','ActivityMinute'), how='left')
min_merge.head()
```

## Data Cleaning

* Now we split the **ActivityMinute** column into two seperate colums **Activity_Date** and **Activity_Minute**

```python
min_merge[['Activity_Date','Activity_Minutes']]= min_merge['ActivityMinute'].str.split(' ', 1, expand=True)
min_merge.head()
```

* Now we will drop the **ActivityMinute** Column as we split it, so we don't need this column now.

```python
min_merge.drop(['ActivityMinute'], axis=1)
```

* Now we will changed the column order

```python
min_merge=min_merge[['Id','Activity_Date','Activity_Minutes','Steps','Calories','Intensity','METs']]
```

```python
min_merge.head()
```

* Round off the calories values

```python
min_merge['Calories'] = np.round(min_merge['Calories'],decimals = 2)
min_merge.head()
```

* As the **Activity_Date** column is consist of Dates but it is shown as object type so we change the Dtyes into date

```python
min_merge['Activity_Date'] = pd.to_datetime(min_merge['Activity_Date'].str.strip())
```
